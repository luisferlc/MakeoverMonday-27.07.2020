```python
df2=df[['Country Name','1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004',
       '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013',
       '2014', '2015', '2016', '2017', '2018', '2019']]
```

```python
df2
```

```python
df3=df2.pivot_table(index=['Country Name']).reset_index()
#Ya se eliminan los rows con puros Nulls
#Si te das cuenta, este dataframe tiene el mismo shape que cuando haces el pivot_table
```

```python
df3
```

```python
id_vars = ['Country Name']
value_vars = ['1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004',
       '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013',
       '2014', '2015', '2016', '2017', '2018', '2019']
var_name = 'Year'
value_name = 'Value'

melt = df3.melt(id_vars=id_vars,value_vars=value_vars,var_name=var_name, value_name=value_name)
melt
```

```python
melt.groupby('Year').agg({'Value':'sum'})
```
